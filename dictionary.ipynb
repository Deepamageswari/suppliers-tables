```python
class sets():
    def taking_input(self):
        self.numbers=set()
        count=0
        h=True
        while (h):
            add_num=int(input())
            self.numbers.add(add_num)
            count+=1
            if count==10:
                h=False
        print("set of numbers",self.numbers)
    def single_digit(self):
        self.Set_single=set()
        count_single=0
        for i in self.numbers:
            if i>=0:
                if i<10:
                    self.Set_single.add(i)
                    count_single+=1
                    self.avg_d=(sum(self.Set_single))/count_single
        print("set of single digits",self.Set_single)
        print("avg of single digits",self.avg_d)
        print("sum of single digits",sum(self.Set_single))
        print("max of single digits",max(self.Set_single))
        print("min of single digits",min(self.Set_single))
        
    def double_digit(self):
        self.Set_double=set()
        count_double=0
        for j in self.numbers:
            if j>=9:
                if j<100:
                    self.Set_double.add(j)
                    count_double+=1
                    self.avg_d=(sum(self.Set_double))/count_double
        print("set of double digits",self.Set_double)
        print("avg of double digits",self.avg_d)
        print("sum of souble digits",sum(self.Set_double))
        print("max of double digits",max(self.Set_double))
        print("min of double digits",min(self.Set_double))
        
    def others(self):
        self.Set_others=set()
        count_others=0
        for k in self.numbers:
            if k>0:
                if k>=100:
                    self.Set_others.add(k)
                    count_others+=1
                    self.avg_o=(sum(self.Set_others))/count_others
        print("set of other digits",self.Set_others)
        print("avg of other digits",self.avg_o)
        print("sum of other digits",sum(self.Set_others))
        print("max of other digits",max(self.Set_others))
        print("min of other digits",min(self.Set_others))
        
class sort(sets):
    def storing(self):
        self.single=list(self.Set_single)
        self.double=list(self.Set_double)
        self.others=list(self.Set_others)
        
    def single_sort(self):
        temp1=0
        for k in range(0,len(self.single)):
            for l in range(k+1,len(self.single)):
                if (self.single[k]>self.single[l]):
                    temp1=self.single[k]
                    self.single[k]=self.single[l]
                    self.single[l]=temp1
        print()
        for k in range(0, len(self.single)):    
            print(self.single[k], end=" ")
            
    def double_sort(self):
        temp2=0
        for m in range(0,len(self.double)):
            for n in range(m+1,len(self.double)):
                if (self.double[m]>self.double[n]):
                    temp2=self.double[m]
                    self.double[m]=self.double[n]
                    self.double[n]=temp2
        print()
        for m in range(0, len(self.double)):    
            print(self.double[m], end=" ")
    
    def others_sort(self):
        temp3=0
        for p in range(0,len(self.others)):
            for q in range(p+1,len(self.others)):
                if (self.others[p]>self.others[q]):
                    temp3=self.others[p]
                    self.others[p]=self.others[q]
                    self.others[q]=temp3
        print()
        for p in range(0, len(self.others)):    
            print(self.others[p], end=" ")

            
    
        
        
        
    
ob=sort()
ob.taking_input()
ob.single_digit()
ob.double_digit()
ob.others()
ob.storing()
ob.single_sort()
ob.double_sort()
ob.others_sort()
        
```
